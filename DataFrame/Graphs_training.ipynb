```python
x1 = ['2010','2011','2012']
y1 = [10, 12.5, 8]

y2 = [13,10,11.5]
```

```python
fig = plt.figure(figsize=(9, 3))
ax1 = plt.subplot2grid((1,1),(0,0))


ax1.plot_date(x1,y1,'-',label ="T_max")
ax1.plot(x1,y2, label = "T_min")



# Rotating the labels
for label in ax1.xaxis.get_ticklabels():
    label.set_rotation(45)

# Create 'Grids'
#ax1.grid(True)#, color= 'g', linestyle = '-',linewidth=2)

#changing the color of the labels
ax1.xaxis.label.set_color('c')
ax1.yaxis.label.set_color('c')

# Changing the intervals between ticks
ax1.set_yticks([8, 10,12,14])

#filling between two axes
ax1.fill_between(x1,y1,10, alpha = 0.3, facecolor='g')

# Spines
ax1.spines['bottom'].set_color('c')
ax1.spines['right'].set_visible(False)
ax1.spines['top'].set_visible(False)
ax1.spines['left'].set_linewidth(5)

# Chnage the color of ticks
ax1.tick_params(axis="x", colors='b')

# Determine horizontal & vertical line

ax1.axvline(x1[1], color="r", linewidth=2)


# Put text on the PLot

font_dict = {'family':'serif', 'color':'darkred','size':15}
ax1.text(x1[1],y2[1], 'The import point', fontdict=font_dict)


# Anotation

#ax1.annotate('Big news!', (x1[2],y2[2]), xytext=(0.8,0,9),
 #            textcoords = 'axes fraction', arrowprops=dict(facecolor='grey'),
#            color='grey')

plt.xlabel("years")
plt.ylabel("prec.")
plt.title("The amount of the rainfals")
plt.legend()
#plt.subplots_adjust(left =0, bottom=0.18, top=0.94, wspace=0.2, hspace=0)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [15, 30, 45, 10]
explode = (0, 0.1, 0, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

### Make a animation plot

```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from matplotlib import style
```

```python
style.use('fivethirtyeight')
```

```python
fig = plt.figure()
ax1=fig.add_subplot(1,1,1)

def animate(i):
    graph_data = open('samplefile.txt', 'r').read()
    lines = graph_data.split('\n')
    xs = []
    ys = []
    for line in lines:
        if len(line) > 1:
            x,y = line.split(',')
            xs.append(x)
            ys.append(y)

    ax1.clear()
    ax1.plot(xs,ys)

ani=animation.FuncAnimation(fig, animate, interval=1000)
plt.show()
```
